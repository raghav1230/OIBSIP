```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Unemployment in India.csv")
```

```python
data.columns
```

```python
data.head(2)
```

```python
data.columns
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
employed = data[' Estimated Employed'].iloc[:50]
employed.head(2)
```

```python
unemployed = data[' Estimated Unemployment Rate (%)'].iloc[:50]
unemployed.head(2)
```

```python
from sklearn.preprocessing import normalize
```

```python
e = normalize([employed])
print(e)
u = normalize([unemployed])
print(u)
```

```python
import numpy as np
```

```python
e =e.flatten()
print(e)

plt.plot(e)
plt.show()


u = u.flatten()
print(u)

plt.plot(u)
plt.show()
```

```python
plt.plot(e, color='red')
plt.plot(u, color='blue')
plt.xlabel("Employed")
plt.ylabel("Unemployed")
plt.show()
```

```python
sns.histplot(x=' Estimated Unemployment Rate (%)',
             hue='Region',
            data=data)

plt.show()
```
